## Dataset Overview

> This is a dataset titled '201902 fordgobikes tripdata'. I will be exploring the relationships between type of user, gender, duration and birth year.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
go_bikes = pd.read_csv('201902_fordgobike_tripdata.csv')
```

```python
go_bikes['start_time'] = pd.to_datetime(go_bikes['start_time'])
go_bikes['end_time'] = pd.to_datetime(go_bikes['end_time'])
```

```python
go_bikes['member_birth_year'] = go_bikes['member_birth_year'].fillna(0)
go_bikes['start_station_id'] = go_bikes['start_station_id'].fillna(0)
go_bikes['end_station_id'] = go_bikes['end_station_id'].fillna(0)

go_bikes = go_bikes.astype({'member_birth_year':'int', 'start_station_id':'int', 'end_station_id':'int' })
```

```python
def reorder_columns(columns, first_cols=['bike_id','start_time','end_time','duration_sec']):
    columns = list(set(columns) - set(first_cols))
    new_order = first_cols + columns
    return new_order
```

```python
reorder_columns(go_bikes, first_cols=['bike_id','start_time','end_time','duration_sec'])
```

```python
reordered_cols = reorder_columns(go_bikes, first_cols=['bike_id','start_time','end_time','duration_sec'])
go_bikes = go_bikes[reordered_cols]
```

```python
go_bikes.bike_id.duplicated().value_counts()
```

```python
#select only values that are not zero
birth_year = go_bikes[go_bikes['member_birth_year'] !=0]
birth_years = birth_year[['member_birth_year','duration_sec']]
```

## Member Gender

> Visualise the distribution of member gender.

```python
#Display gender distribution
#first we need to get data from the 'member_gender' column
pie_data = go_bikes.member_gender.value_counts()
pie_data

#Plot the pie chart

pie, ax = plt.subplots(figsize=[10,6])
labels = pie_data.keys()
explode = (0.02,0.02,0.09)
plt.pie(x = pie_data, autopct = "%.1f%%", explode = explode, labels = labels, pctdistance = 0.8)
plt.title("Member Gender", fontsize = 14);
plt.legend(["Male", "Female", "Other"]);
```

## User type chart
>Looking at the different user types using a piechart. There are 89.2% of subscribers as users.

```python
#Display user type distribution

user_type_data = go_bikes.user_type.value_counts()
user_type_data

#Plot user type pie chart

pie, ax = plt.subplots(figsize=[10,6])
labels = user_type_data.keys()
explode = (0.02,0.02)
plt.pie(x = user_type_data, autopct = "%.1f%%", explode = explode, labels = labels, pctdistance = 0.8)
plt.title("User Type", fontsize = 14);
plt.legend(["Subscriber", "Customer"]);  
```

## Birth Year and Duration

> In this scatter plot we see that users born between 1980 and 2000 spend more time on trips.

```python
#plot member birth year vs duration in seconds with a colormap of different bike ids

plt.scatter(data = birth_year, x = 'member_birth_year', y = 'duration_sec', c='bike_id')
plt.xlabel('memeber_birth_year')
plt.ylabel('duration_sec')
plt.title('Birth year VS Duration through Bike id')
plt.colorbar();
```

## Birth Year and Gender
> In the violin and box plot here we are looking at member birth year and gender. The violin plot is showing that female and male users are the same across user types but the distribution of user type other is more wide in subcribers vs customers.


```python
plt.figure(figsize = [16, 5])
base_color = sb.color_palette()[0]

#Left plot, violin plot 
plt.subplot(1, 2, 1)
sb.violinplot(data = birth_year, x = 'member_gender', y = 'member_birth_year',palette = 'Set1', inner = 'quartile', split = True, hue = 'user_type')
plt.xticks(rotation=90)
plt.xlabel('member_gender')
plt.ylabel('member_birth_year')
plt.title("Birth year vs Gender")

# RIGHT plot: box plot
plt.subplot(1, 2, 2)
sb.boxplot(data = birth_year, x = 'member_gender', y = 'member_birth_year', palette = 'Set2', hue = 'user_type')
plt.xticks(rotation=90)
plt.title("Birth year vs Gender");
```

## Gender and Duration
> Here we look at the duration of each trip by gender split between customer and subscriber. What this bar plot shows is that on all genders customers have a higher duration than subscribers with other gender spending more time than female and male.

```python
#plotting member gender vs duration grouped by the user type
ax = sb.barplot(data = birth_year, x = 'member_gender', y = 'duration_sec', hue = 'user_type')
plt.title('Member gender VS Duration through user type');
```

## Duration, Gender and BikeID
> Bike IDs that are numbered from 4000 and above are the ones being used for longer durations on trips. Members born after 1980 and 2000 have the widest distribution of duration spent on trips and most of the users are subscribers.

```python
#plotting a pairgrid looking at member birth year, bike id and duration grouped by user type
grid = sb.PairGrid(data = birth_year, vars = ['member_birth_year', 'duration_sec', 'bike_id'], hue = 'user_type')
grid.map_diag(sb.histplot)
grid.map_offdiag(sb.scatterplot)
grid.add_legend();
```
